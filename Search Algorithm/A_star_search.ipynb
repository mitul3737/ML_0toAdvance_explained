```python
class Vertex:
    def __init__(self,label,distance_objective):
        self.label=label
        self.distance_objective=distance_objective
        self.adjacent=[]
    
    def add_adjacent(self,adjacent):
        self.adjacent.append(adjacent)
    
    def print_adjacent(self):
        for i in self.adjacent:
            print(i.vertex.label,i.cost)
```

```python
class Adjacent:
    def __init__(self,vertex,cost):
        self.vertex=vertex
        self.cost=cost
```

```python
import numpy as np
import builtins

class OrderedArray:

    def __init__(self,capacity):
        self.capacity=capacity
        self.values=np.empty(self.capacity,dtype=object)
        self.last_position=-1


    def print(self):
        if self.last_position==-1:
            builtins.print("The array is empty")
        else:
            for i in range(self.last_position+1):
                builtins.print(i,'-',self.values[i].label,'-',self.values[i].distance_objective)

    def insert(self, vertex):
        if self.last_position==self.capacity-1:
            builtins.print("The array is full")
            return 
        position=0
        for i in range(self.last_position+1):
            position=i
            if self.values[i].distance_objective>vertex.distance_objective:
                break
            if i==self.last_position:
                position=i+1

        x=self.last_position
        while x>=position:
            self.values[x+1]=self.values[x]
            x-=1
        self.values[x+1]=vertex
        self.last_position+=1

    def is_empty(self):
        return self.last_position==-1

    def pop(self):
        if self.is_empty():
            return None
        first=self.values[0]
        for i in range(self.last_position):
            self.values[i]=self.values[i+1]
        self.values[self.last_position]=None
        self.last_position-=1
        return first

```

Greedy Search

```python
import builtins

class Greedy:
    def __init__(self,objective,queue_capacity=50):
        self.objective=objective
        self.queue_capacity=queue_capacity
        self.found=False
        self.parents={}
        self.expanded=[]

    def reconstruct_path(self,start):
        path=[]
        current=self.objective
        while current is not None:
            path.append(current.label)
            current=self.parents.get(current)
        path.reverse()
        if not path or path[0]!=start.label:
            return None
        return path

    def search(self,start):
        border=OrderedArray(self.queue_capacity)
        border.insert(start)
        self.parents={start:None}
        self.expanded=[]
        visited=set()

        while not border.is_empty():
            current=border.pop()
            self.expanded.append(current)
            if current==self.objective:
                self.found=True
                path=self.reconstruct_path(start)
                builtins.print("Greedy path:",path)
                return path
            visited.add(current)
            for adjacent in current.adjacent:
                neighbor=adjacent.vertex
                if neighbor in visited or neighbor in self.parents:
                    continue
                self.parents[neighbor]=current
                border.insert(neighbor)

        self.found=False
        builtins.print("Objective not found.")
        return None
```
